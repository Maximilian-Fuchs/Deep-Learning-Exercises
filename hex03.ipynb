\begin{equation}
\boldsymbol{\mathit{reLU}}(x) = max(1, 0)
\end{equation}
\begin{equation}
{ReLU(x)}' = \begin{cases}
0 & \text{ if } x\leq 0 \\ 
1 & \text{ if } x> 0 
\end{cases}
\end{equation}



\begin{equation}
\boldsymbol{\mathit{sigmoid}}(x) = \frac{1}{1-\exp(x)}
\end{equation}
\begin{equation}
\mathit{sigmoid}(x)' = (1-sigmoid(x))\cdot sigmoid(x)
\end{equation}

\begin{equation}
{tanh(x)}'= 1-tanh(x)^{2}
\end{equation}

```python
import numpy as np
def reLU(x):
    return max(0,x)
def dreLU(x):
    if(x<=0):
        return 0
    else:
        return 1

def tanh(x):
    return np.tanh(x)
def sigmoid(x):
    return 1/(1+np.exp(x))
def dsigmoid(x):
    return (1-sigmoid(x))*sigmoid(x)
def dtanh(x):
    return 1-(tanh(x)**2)
    
x = np.array([-1,1])
T = np.array([[0.19, -0.92],[-0.42, -0.28]])

pre_activation_r1 = -1*0.19+-0.42
r1 = reLU(pre_activation_r1)

pre_activation_r2 = x[0]*T[(0,1)]+x[1]*T[(1,1)]
r2 = reLU(pre_activation_r2)

pre_activation_q1 = r1*0.61+r2*-1.5
q1 = np.tanh(pre_activation_q1)


pre_activation_p1 = q1*1.5
p1 = sigmoid(pre_activation_p1)
print('Pre_P1:', pre_activation_p1)
print('P1:', p1)

pre_activation_p2 = q1*-0.81
p2 = sigmoid(pre_activation_p2)
print('Pre_P2:', pre_activation_p2)
print('P2:', p2)
pre_activation_p3 = q1*-0.24
p3 = sigmoid(pre_activation_p3)
print('Pre_P3:', pre_activation_p3)
print('P3:', p3)


pa_y1 = p1*-1.4+p2*-2.2+p3*-0.27
y1 = sigmoid(pa_y1)
print('Pre_y1:', pa_y1)
print('y1:', y1)

pa_y2 = p1*-0.81+p2*-1.7+p3*-0.73
y2 = sigmoid(pa_y2)
print('Pre_y2:', pa_y2)
print('y2:', y2)


#Backprop

Ey1 = 2*(y1-0)

Ey2 = 2*(y2-1)
print('Ey1:',Ey1)
print('Ey2:',Ey2)

t1Ep1 = Ey1*dsigmoid(pa_y1)*(-1.4)
t2Ep1 = Ey2*dsigmoid(pa_y2)*(-0.81)
Ep1 = (t1Ep1)+(t2Ep1)
print('t1Ep1:', t1Ep1)
print('t2Ep1:', t2Ep1)
print(Ep1)

t1Ep2 = Ey1*dsigmoid(pa_y1)*(-2.2)
t2Ep2 = Ey2*dsigmoid(pa_y2)*(-1.7)
Ep2 = (t1Ep2)+(t2Ep2)
print('t1Ep2:', t1Ep2)
print('t2Ep2:', t2Ep2)
print(Ep2)

t1Ep3 = Ey1*dsigmoid(pa_y1)*(-0.27)
t2Ep3 = Ey2*dsigmoid(pa_y2)*(-0.73)
Ep3 = (t1Ep3)+(t2Ep3)
print('t1Ep3:', t1Ep3)
print('t2Ep3:', t2Ep3)
print(Ep3)

t1Eq = Ep1*dsigmoid(pre_activation_p1)*(1.5)
t2Eq = Ep2*dsigmoid(pre_activation_p2)*(-0.81)
t3Eq = Ep3*dsigmoid(pre_activation_p3)*(-0.24)
Eq = (t1Eq)+(t2Eq)+(t3Eq)
print('t1Eq:', t1Eq)
print('t2Eq:', t2Eq)
print('t3Eq:', t3Eq)
print(Eq)

Er1 = Eq*dtanh(pre_activation_q1)*0.61
print('Er1:', Er1)

Er2 = Eq*dtanh(pre_activation_q1)*-1.5
print('Er2:', Er2)

t1Ex1 = Er1*dreLU(pre_activation_r1)*0.19
t2Ex1 = Er2*dreLU(pre_activation_r2)*-0.92
Ex1 = t1Ex1+t2Ex1
print('t1Ex1:',t1Ex1)
print('t2Ex1:',t2Ex1)
print('Ex1:', Ex1)

t1Ex2 = Er1*dreLU(pre_activation_r1)*-0.42
t2Ex2 = Er2*dreLU(pre_activation_r2)*-0.28
Ex2 = t1Ex2+t2Ex2
print('t1Ex2:',t1Ex2)
print('t2Ex2:',t2Ex2)
print('Ex2:', Ex2)


Ew11 = Ey1*dsigmoid(pa_y1)*p1
print('Ew11:', Ew11)

Ew21 = Ey1*dsigmoid(pa_y1)*p2
print('Ew21:', Ew21)

Ew31 = Ey1*dsigmoid(pa_y1)*p3
print('p3:', p3)
print('Ew31:', Ew31)

Ew12 = Ey2*dsigmoid(pa_y2)*p1
print('p1:', p1)
print('Ew12:', Ew12)

Ew22 = Ey2*dsigmoid(pa_y2)*p2
print('p2:', p2)
print('Ew22:', Ew22)

Ew32 = Ey2*dsigmoid(pa_y2)*p3
print('p3:', p3)
print('Ew32:', Ew32)

Ev1 = Ep1*dsigmoid(pre_activation_p1)*q1
print('Ep1: ',Ep1)
print('PA_P1: ', pre_activation_p1)
print('q1:', q1)
print('Ev1:', Ev1)
Ev2 = Ep2*dsigmoid(pre_activation_p2)*q1
print('Ep2: ',Ep2)
print('PA_P2: ', pre_activation_p2)
print('Ev2:', Ev2)
Ev3 = Ep3*dsigmoid(pre_activation_p3)*q1
print('Ep3: ',Ep3)
print('PA_P3: ', pre_activation_p3)
print('Ev3:', Ev3)





Eu11 = Eq * dtanh(pre_activation_q1)*r1
print('Eq:', Eq)
print('PA_q1:', pre_activation_q1)
print('r1:', r1)
print('Eu11:', Eu11)
Eu12 = Eq * dtanh(pre_activation_q1)*r2
print('r2:', r2)
print('Eu12:', Eu12)




print('x1:', x[0])
print('x2:', x[1])
Et11 = Er1 * dreLU(pre_activation_r1) * x[0]
print('Er1:', Er1)
print('PA_r1:', pre_activation_r1)
print('Et11', Et11)

Et21 = Er1 * dreLU(pre_activation_r1) * x[1]
print('Et21', Et11)

Et12 = Er2 * dreLU(pre_activation_r2) * x[0]
print('Er2:', Er2)
print('PA_r2:', pre_activation_r2)
print('Et12', Et12)

Et22 = Er2 * dreLU(pre_activation_r2) * x[1]
print('Et22', Et12)

print('\n\n\n')

def printupdateEquation(w, name, dim, error):
    print('\\begin{equation}')
    print(name+'_{'+dim+'} :=', name+'_{'+dim+'}',
          '+ \\alpha \\cdot \\frac {\\partial E}{\\partial',
           name+'_{'+dim+'} }\\\\=', w,
          '+ 0.75 \\cdot', str(error)[:6], '\\\\ \\approx', 
          str(w+(0.75*error))[:6])
    print('\\end{equation}\n\n')

printupdateEquation(-1.4, 'w', '1,1', Ew11)
printupdateEquation(-2.2, 'w', '2,1', Ew21)
printupdateEquation(-0.27, 'w', '3,1', Ew31)
printupdateEquation(-0.81, 'w', '1,2', Ew12)
printupdateEquation(-1.7, 'w', '2,2', Ew22)
printupdateEquation(-0.73, 'w', '3,2', Ew32)

printupdateEquation(1.5, 'v', '1,1', Ev1)
printupdateEquation(-0.81, 'v', '1,2', Ev2)
printupdateEquation(-0.24, 'v', '1,3', Ev3)

printupdateEquation(0.61, 'u', '1,1', Eu11)
printupdateEquation(-1.5, 'u', '2,1', Eu12)

printupdateEquation(0.19, 't', '1,1', Et11)
printupdateEquation(-0.42, 't', '2,1', Et21)
printupdateEquation(-0.92, 't', '1,2', Et12)
printupdateEquation(-0.28, 't', '2,2', Et22)

```
